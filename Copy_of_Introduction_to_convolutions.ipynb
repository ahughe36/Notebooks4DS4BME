## Convolutions

Given two vectors, $a$ and $b$, a discrete convolution $c$ is defined in the terms of element $n$ as: 
$$
c_n = \sum_{m=-\infty}^{m=\infty} a_m b_{n-m}
$$
There is also a functional definition as $c(k) = \int_{-\infty}^\infty f(t)g(k-t)dt$. 

What does convolution do? It's like smearing $a$ through $b$. So, convolutions are typically used for smoothing. Wikipedia has a great animation of a convolution [here](https://en.wikipedia.org/wiki/Convolution). It's important to note that a good way to perform convolutions is to take the Fourier transforms of $a$ and $b$, multiply in Fourier space, then invert; that's the convolution.

Of course, we don't have vectors from $-\infty$ to $\infty$. 


```
import numpy as np
import matplotlib.pyplot as plt
```

```
## create a signal
x = np.random.normal(size = 50)
plt.plot(x)
plt.show()
```

```
## Create a kernel
k = np.zeros(10) + 1
k = k / k.size
```

```
y = np.convolve(x, k)

plt.plot(x)
plt.plot(y)
plt.show()
```

```
n = x.size
m = k.size
xzp = np.pad(x, (0, m-1))
kzp = np.pad(k, (0, n-1))
y2 = np.fft.ifft(np.fft.fft(xzp) * np.fft.fft(kzp)).real
plt.plot(x)
plt.plot(y2)
plt.show()
```
